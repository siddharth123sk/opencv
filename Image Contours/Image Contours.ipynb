```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

#read the image
image = cv2.imread("hand.jpg")

#making a copy of the image
image_copy = np.copy(image)

#change the color from BGR to GRB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
#Producing a binary image for finding contourts around the hand
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

#creating bnary threshold image
retval, binary = cv2.threshold(gray, 245, 255, cv2.THRESH_BINARY_INV)
#takes gray image, isolates white pixels and turns them black

plt.imshow(binary, cmap ="gray")
```

```python
## Finding contours from threshold image

contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#take input as a gray image, contour retrival mode which here is a tree, followed by contour approximation method which here is a simple chain

#draw the contours on a copy of the original image
image_copy2 = np.copy(image_copy)
all_contours = cv2.drawContours(image_copy2, contours, -1, (0,255,0), 2)
#this takes input as our image followed by list of contours, the which contours to display, -1 means all, then colour(0,255,0) and size(2) of the contours

plt.imshow(all_contours)
```
