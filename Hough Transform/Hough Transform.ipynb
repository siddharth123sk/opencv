```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('hough transform.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

lower = 50
upper = 100

edges = cv2.Canny(gray, lower, upper)

plt.imshow(edges, cmap = 'gray')
```

## Finding lines through Hough transform

```python
#defining the hough transform parameters

rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 100
max_line_gap = 5

#finding lines using hough transform

lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

line_image = np.copy(image_copy)

#iterating over the output lines and drawing lines on the image copy

for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(line_image,(x1,y1), (x2,y2), (255,0,0),5)
        
plt.imshow(line_image)
```
