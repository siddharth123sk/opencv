```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open("examples/english.png")
im
```

```python
import easyocr
reader = easyocr.Reader(['en'])
```

```python
bounds = reader.readtext('examples/english.png')
bounds
```

```python
def drawing_boxes(image, bounds, color = 'green', width = 2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill = color, width = width)
    return image

drawing_boxes(im, bounds)
```
