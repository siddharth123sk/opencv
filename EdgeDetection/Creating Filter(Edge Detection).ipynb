```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('test.png')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

# Custom Kernel

```python
# creating edge detection kernel
# also called sobel filer #vertical edge detection
sobel_x = np.array([[ -1, 0, 1],
                   [ -2, 0, 2],
                   [ -1, 0, 1]])

filtered_image = cv2.filter2D(gray, -1, sobel_x)

plt.imshow(filtered_image, cmap = 'gray')
```

## Creating binary image

```python
retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap = 'gray')
```

## Gaussian Blur

```python
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)

f, (ax1, ax2) = plt.subplots(1,2, figsize  = (20,10))

ax1.set_title('original gray')
ax1.imshow(gray, cmap = 'gray')

ax2.set_title('blurred image')
ax2.imshow(gray_blur, cmap = 'gray')
```

## Comparing the effects of a low pass filter (Gaussian Blur) with high pass filter(Sobel Edge Detection) and a normal image

```python
sobel_x = np.array([[ -2, 0, 2],
                   [ -1, 0, 1],
                   [ -2, 0, 2]])

filtered = cv2.filter2D(gray, -1, sobel_x)

filtered_blurred = cv2.filter2D(gray_blur, -1, sobel_x)

f, (ax1, ax2) = plt.subplots(1,2, figsize = (20,10))

ax1.set_title('original gray')
ax1.imshow(filtered, cmap = 'gray')

ax2.set_title('blurred image')
ax2.imshow(filtered, cmap = 'gray')
```

## Implementing canny edge detection

```python
lower = 120
upper = 240

edges = cv2.Canny(gray, lower, upper)

plt.imshow(edges, cmap = 'gray')
```

```python
wide = cv2.Canny(gray, 30 ,100)
tight = cv2.Canny(gray, 180, 240)

f, (ax1, ax2) = plt.subplots(1,2, figsize  =(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap = 'gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap = 'gray')
```
